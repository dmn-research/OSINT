```python
# Make the API request
url = "https://senate.texas.gov/coi.php"
response = requests.get(url)
data = response.json()

# Extract the links from the response
links = []
for item in data:
    if "website" in item:
        links.append(item["website"])

# Print the extracted links
for link in links:
    print(link)
```
